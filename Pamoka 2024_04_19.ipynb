```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import requests as rq
import warnings
warnings.filterwarnings('ignore')
```

```python
# nuskaitykite auto.csv, ir pridėkite prie šios lentelės tokių stulpelių : 
# gamintojo trys pirmos raidės didžiosiomis, suskaičiuokite, kiek kiekvienas gamintojas turėjo skirtingų automobilių modelių, 
# ADVANCED - pavaizduokite tai grafiškai. Grafiko pobūdį ir kaip tai padarysite - renkatės patys. Hint - .groupby().reset_index() gali duoti lentelę

auto = pd.read_csv('auto.csv', sep='|')
auto_mini = auto[['gamintojas', 'modelis']]
auto_mini.head()

gamintojai = ['Mas','Lan', 'Tes', 'Por', 'Ast']

def gamintojoKodas(raides):
    trys_raides = raides[:3]
    return trys_raides

auto_mini['gamintojo_kodas']=auto_mini['gamintojas'].apply(gamintojoKodas)
auto_mini['modelis2']=auto_mini['modelis']

auto_mini2= auto_mini[['gamintojo_kodas']].groupby(by='gamintojo_kodas').value_counts().reset_index()
# auto_mini2= auto_mini[['gamintojo_kodas','modelis2']].groupby(by='gamintojo_kodas').value_counts().reset_index() - cia klaida, nes nereikia modelis2 itraukti

# auto_mini2.query('gamintojo_kodas == @gamintojai').value_counts().unstack().plot(kind='bar')
```

```python
auto_mini.query('gamintojo_kodas in @gamintojai').groupby(by=['gamintojas', 'modelis']).count().reset_index().groupby(by='gamintojas').count().reset_index()[['gamintojas','modelis']]

auto_mini.query('gamintojo_kodas == @gamintojai').value_counts().unstack().plot(kind='bar') # cia reikia pataisyti
```
